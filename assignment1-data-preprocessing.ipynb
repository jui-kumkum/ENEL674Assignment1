```python
import control as ct
import numpy as np
import scipy as sp  
import matplotlib.pyplot as plt
import pickle as pkl

dat_file_load = open('assignment1.pkl','rb')
tt, uu, yy = pkl.load(dat_file_load)
tt_u = np.arange(0,10,0.01)

num_exp = yy.shape[1]
N = yy.shape[0]
print('Data length=',N)
print('number of experiments:', num_exp)

ii=1
fig,ax = plt.subplots(1)
ax.plot(tt_u,uu[:,ii],label='input')
ax.plot(tt[0:-1,ii], yy[0:-1,ii],label='original')
ax.set_title('Raw Data')
ax.set_xlabel('Time (s)')
ax.legend()





 
```

```python
yy_interpolated = np.interp(tt_u,tt[0:-1,ii],yy[0:-1,ii])
fig,ax = plt.subplots(1)
ax.plot(tt[0:-1,ii], yy[0:-1,ii],label='original')
ax.plot(tt_u, yy_interpolated,label='interpolated')
ax.set_title('Output Data')
ax.set_xlabel('Time (s)')
ax.legend()



```

```python

t0 = 100
yy_detrend = sp.signal.detrend(yy_interpolated[t0:N-1])
yy_processed = yy_detrend

fig,ax = plt.subplots(1)
ax.plot(tt[0:-1,ii], yy[0:-1,ii],label='original')
ax.plot(tt_u, yy_interpolated,label='interpolated')
ax.plot(tt_u[t0:N-1], yy_detrend,label='detrend')
ax.set_title('Output Data')
ax.set_xlabel('Time (s)')
ax.legend()
```
